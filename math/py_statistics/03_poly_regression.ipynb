# Polynomial Regression
- using numpy polynomial model

## Reference
[Linear Regression in Python using numpy + polyfit ](https://data36.com/linear-regression-in-python-numpy-polyfit/)

```python
import numpy
import matplotlib.pyplot as plt
x = [1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900]
y = [1100, 1150, 1200, 1275, 1375, 1550, 1700, 1800, 1900]
plt.scatter(x, y)
plt.show()
```

```python
mymodel = numpy.poly1d(numpy.polyfit(x, y, 2))
myline = numpy.linspace(1100, 1900, 800)
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show() 
```

```python
from sklearn.metrics import r2_score
print(r2_score(y, mymodel(x))) 
```
